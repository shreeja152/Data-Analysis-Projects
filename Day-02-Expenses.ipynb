```python
# Daily expences in rupees

expences = [ 123, 765, 901, 15, 2003, 2]

total_expences = sum(expences)
average_expences = total_expences/ len(expences)        # len - returns the number of elements in a list (or string, tuple, etc.).

print ("Expences : ", expences)
print ("Total Expences : " , total_expences)
print ("Average expences : ", average_expences)
```
